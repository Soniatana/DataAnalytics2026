## Working with custom functions

```python
def favorite_things(name, movie, music):
    print("Name:", name)
    print("Favorite movie:", movie)
    print("Favorite musician/band:", music)
```

```python
favorite_things("Sonia", "Black Panther", "Burna Boy")
```

```python
def why_im_here(purpose):
    return purpose   
```

```python
print(why_im_here("I joined Year Up to improve my technical and professional skills in data analytics."))
```

```python
def favorite_place(place, reason):
    print(f"My favorite place is {place} because {reason}.")
```

```python
favorite_place("Charlotte", "it gives me opportunities to grow professionally")
```
