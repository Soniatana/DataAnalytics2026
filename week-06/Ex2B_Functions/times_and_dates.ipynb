# Lab 2 of Exercise 2.B - Working with datetime functions

```python
import datetime
```

# Today

```python
today = datetime.datetime.now()

print(today)
```

```python
formatted_today = today.strftime("Today is %A, %B %d, %Y")

print(formatted_today)
```

# Next 90 Days

```python
ninety_d = today + datetime.timedelta(days=90)
print(ninety_d)
```

```python
future_date = ninety_d.strftime("90 days from today is %B %d, %Y")

print(future_date)
```

# Time for Dinner

```python
dinner_time = datetime.time(18, 30)
print(dinner_time)
```

```python
formatted_dinner = dinner_time.strftime("let's meet for dinner at %I:%M %p")
print(formatted_dinner)
```
