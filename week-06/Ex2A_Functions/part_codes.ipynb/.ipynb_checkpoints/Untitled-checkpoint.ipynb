# Exercise 2.A Using built-in functions
## Lab 2. Parsing strings – part codes

```python
def get_supplier_code(part_code):
    return part_code.split(":")[0]
```

```python
print(get_supplier_code("ACME:123-L"))
```

```python
def get_product_num(part_code):
    return part_code.split(":")[1].split("-")[0]
```

```python
print(get_product_num("ACME:123-L"))
```

```python
def get_size(part_code):
    return part_code.split("-")[1]
```

```python
print(get_size("ACME:123-L"))
```

```python
part1 = "ACME:123-L"
part2 = "DI:12345-M"
part3 = "ACE:1-12"
```

```python
print(
    f"For {part1}, the supplier is {get_supplier_code(part1)}. "
    f"Part number {get_product_num(part1)} is available in size {get_size(part1)}."
)
```

```python
print(
    f"For {part2}, the supplier is {get_supplier_code(part2)}. "
    f"Part number {get_product_num(part2)} is available in size {get_size(part2)}."
)
```
