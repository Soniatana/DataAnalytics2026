# Exercise 1.A NumPy: Manipulating Arrays

```python
import numpy as np
```

## Looping Through Arrays

```python
arrA = np.array([
    [11, 12, 13],
    [14, 15, 16]
])
print(arrA)
```

```python
for x in arrA:
    print(x * 2)
```

```python
for x in arrA:
    for y in x:
        print(y * 2)
```

first loop goes through each row in the array and multiplies the entire row by 2. The second loop goes through each individual value inside the rows and multiplies each number by 2 separately.

```python
for x in arrA:
    loop_result = x * 2
    print(loop_result)
 # This is different because the result is being stored in a variable before printing.
```

## Joining and Stacking Arrays

```python
joined_array_1 = np.concatenate((arrA, arrA))
print(joined_array_1)
```

```python
joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)
```

```python
stacked_array_1 = np.stack((joined_array_1, joined_array_1))

print(stacked_array_1)
# np.stack combines arrays into a new dimension.
```

```python
another_2 = np.concatenate((joined_array_1, arrA), axis=1)

print(another_2)
# The error happens because the arrays do not have the same number of rows.
```

```python
vertical_stack = np.vstack((arrA, arrA))
print(vertical_stack)
# vstack joins arrays vertically (top and bottom)
```

```python
depth_stack = np.dstack((arrA, arrA))
```

## Splitting Arrays

```python
splitsky = np.array_split(joined_array_1, 2)

print(splitsky)

# The array was split into 2 arrays.
# Each split array still has 2 dimensions
```

```python
spliterino = np.array_split(joined_array_1, 3)

print(spliterino)

# The array was split into 3 arrays.
# Each split array still has 2 dimensions
```

## Searching and Replacing Array Values

```python
result_1 = np.where(arrA % 2 == 0, arrA, 'odd')

print(result_1)
```

## Sorting and Filtering Arrays

```python
arrB = np.array([
    [7, 6],
    [13, 5],
    [7, 3]
])
arrB_sorted = np.sort(arrB)
print(arrB_sorted)
```

```python
print(arrB)

z = arrB < 7

arrC = arrB[z]

print(arrC)
```

```python
y = arrB >= 7

arrD = arrB[y]

print(arrD)
```
