# Exercise 3.A Getting familiar with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.show()
```

```python
plt.plot(x, y, marker='o')

plt.title("Example Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()
```

```python
plt.plot(x, y, linestyle='--', color='red')
plt.show()
```

```python
plt.plot(x, y, marker='s', markersize=10, color='green')
plt.show()
```

```python
plt.plot(x, y, marker='^', color='blue')
plt.show()
```

```python
plt.plot(x, y, marker='*', color='purple')
plt.show()
```
