## Generating Random Numbers

```python
r_1 = np.random.randint(1, 21, size=2)
print(r_1)
```

```python
r_2 = np.random.randint(1, 21, size=(11, 2))
print(r_2)
```

```python
r_3 = np.random.rand(2, 1)
print(r_3)
```

```python
r_4 = np.random.randint(20, 101, size=(4, 5))

print(r_4)
```

```python
r_5 = np.random.rand(4, 5)

print(r_5)
```

## Generating Random Choices from an Array

```python
students = np.array([
    "Sonia",
    "Alex",
    "Maria",
    "John",
    "Sarah",
    "David",
    "Linda",
    "Chris",
    "James",
    "Sophia",
    "Daniel",
    "Emma"
])
```

```python
call_on = np.random.choice(students)

print(call_on)
```

```python
breakout_groups = np.random.choice(students, size=(3, 4))

print(breakout_groups)
# This may not be the best way to assign groups because some students could appear more than once while others may not appear at all.
```

```python
call_on_weighted = np.random.choice(
    students,
    size=5,
    p=[0.30, 0.10, 0.10, 0.10, 0.10, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.00]
)

print(call_on_weighted)

# Sonia has a 30% chance of being selected.
# Emma has a 0% chance of being selected.
# The other students share the remaining probabilities
```

## Generating a Normal Distribution

```python
bell_curve = np.random.normal(loc=10, scale=5, size=(4, 5))

print(bell_curve)
```
