## Sample Arrays

```python
r_1 = np.random.randint(1, 21, size=2)

r_2 = np.random.randint(1, 21, size=(1, 2))

r_3 = np.random.rand(2, 1)

r_4 = np.random.randint(20, 101, size=(4, 5))

r_5 = np.random.rand(4, 5)

print(r_1)
print(r_2)
print(r_3)
print(r_4)
print(r_5)
```

## Permutations

```python
print(r_4)
```

```python
np.random.shuffle(r_4)

print(r_4)
```

The shuffle function rearranged the rows in the array. The values inside each row stayed together, but the order of the rows changed.

## Mathematical Operations

```python
print(r_1)
print(r_1.shape)

print(r_2)
print(r_2.shape)

result_1 = np.add(r_1, r_2)

print(result_1)
print(result_1.shape)
```

```python
print(r_1)
print(r_1.shape)

print(r_3)
print(r_3.shape)

result_2 = np.add(r_1, r_3)

print(result_2)
print(result_2.shape)
```

```python
print(r_2)
print(r_2.shape)

print(r_4)
print(r_4.shape)

result_3 = np.add(r_2, r_4)

print(result_3)
print(result_3.shape)

# This error happens because the arrays have incompatible shapes.
# r_2 has shape (1,2) while r_4 has shape (4,5),
# so NumPy cannot broadcast them together.
```

```python
np.add(r_1, r_4)
# r_1 and r_4 cannot be added because their shapes are incompatible for broadcasting.
```

```python
print(r_4)

print(np.add(np.array([1]), r_4))
```

```python
print(np.add(np.array([[1]]), r_4))
```

```python
print(np.add(np.array([[1]]), r_4))
```

```python
print(np.add(np.array([[1],[1]]), r_4))
# This error happens because the shapes do not align correctly for broadcasting.
```

Broadcasting allows NumPy to perform operations on arrays with different shapes when their dimensions are compatible. Smaller arrays are expanded to match larger arrays. The resulting shape depends on how the dimensions align during the operation.

## Grocery Problem

```python
bag1 = np.array([2, 3, 4, 0])

bag2 = np.array([4, 0, 1, 2])

print(bag1)
print(bag2)
```

```python
bag1_total = np.sum(bag1)
bag2_total = np.sum(bag2)

print(f"The first bag contains {bag1_total} items, and the second bag contains {bag2_total} items.")
```

```python
category_totals = np.add(bag1, bag2)

print(category_totals)

print(f"There are {category_totals[0]} vegetables, {category_totals[1]} fruits, {category_totals[2]} boxes of snacks, and {category_totals[3]} beverages.")
```

## Teamwork

```python
ali = np.array([5, 4, 6, 4, 2])

print(ali)
```

```python
ben = np.array([0, 5, 5, 3, 4])

print(ben)
```

```python
carla = np.array([3, 6, 0, 3, 4])

print(carla)
```

```python
all_knocks = np.array([ali, ben, carla])

print(all_knocks)
```

```python
person_totals = np.sum(all_knocks, axis=1)

group_total = np.sum(all_knocks)

print(person_totals)

print(group_total)
```

```python
daily_totals = np.sum(all_knocks, axis=0)

rolling_total = np.cumsum(daily_totals)

estimated_votes = rolling_total / 2

print(estimated_votes)
```
