## Series

```python
aa = [1, 2, 3, 4]
ab = (5, 6, 7, 8)
ac = {9, 10, 11, 12}
ad = {
    'first': 13,
    'second': 14,
    'third': 15,
    'fourth': 16
}

print(aa)
print(ab)
print(ac)
print(ad)
```

```python
ser1 = pd.Series(aa)

ser2 = pd.Series(ab)

ser3 = pd.Series(list(ac))

ser4 = pd.Series(ad)

print(ser1)

print(ser2)

print(ser3)

print(ser4)
```

```python
ser5 = pd.Series(17, 18, 19, 20)

# TypeError occurs because pd.Series() expects one main data argument,
# not multiple separate integers.
```
